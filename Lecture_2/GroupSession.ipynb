```python
import nltk
nltk.download("brown")
```

```python
from nltk.corpus import brown
wordlist = brown.tagged_words()
wordlist[:20]
```

```python
import collections
d={}
for word,tag in wordlist:
    tag= tag.split("-")[0].split("+")[0]
    if tag not in d:
        d[tag]= collections.defaultdict(int)
    d[tag][len(word)]+=1
print(d["NN"])
print(d["AT"])
print(d["NP"])
```

```python
import pandas as pd
df = pd.DataFrame.from_dict(d)
df
```

```python
df = df.fillna(0)
df
```

```python
p= df.sum()
p.sort_values(ascending=False, inplace=True)
p[0:5]
```
